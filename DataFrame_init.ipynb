```python
import pandas as pd 
import numpy as np
from datetime import datetime
```

***current_date*** = Váriavel, convertida em string, que armazena a data da última atualização do programa. 

***date*** = lista de datas. 

```python
current_date = datetime.now().strftime('%Y%m%d')
date = pd.date_range(current_date, periods=6, freq='D')
print(date)
```

Os índices representam as linhas

```python
df = pd.DataFrame(np.random.randn(6,4), index = date, columns=list('ABCD'))
df
```

```python
df.head(3) # Aparecer as 3 primeiras linhas 
df.tail(3) #Aparecer as 3 últimas linhas
```

Inserir uma nova coluna em um DataFrame. 

**Observação:** As colunas devem ter a mesma conquidade de linhas que as outras.

```python
df['E'] = df['A'] * df['C']
df
```

Concatenar DataFrames

```python
df1 = pd.DataFrame({'A': ['a0', 'a1', 'a2', 'a3'],
                   'B': ['b0', 'b1', 'b2', 'b3'],
                   'C': ['c0', 'c1', 'c2', 'c3'],
                   'D': ['d0', 'd1', 'd2', 'd3']}, 
                   index =[0,1,2,3])

df2 = pd.DataFrame({'A': ['a4', 'a5', 'a6', 'a7'],
                   'B': ['b4', 'b5', 'b6', 'b7'],
                   'C': ['c4', 'c5', 'c6', 'c7'],
                   'D': ['d4', 'd5', 'd6', 'd7']},
                   index = [i for i in range(4,8)])

df1.append(df2)
```

```python
set_df = pd.concat([df1, df2], keys=['FistFrame', 'SecondFrame'])
set_df
```
