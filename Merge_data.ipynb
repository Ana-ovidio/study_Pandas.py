```python
import pandas as pd
import numpy as np
```

```python
# Cadastro da loja a
cadastro_a = {'Id': ['AA2930','BB4563','CC2139','DE2521','GT3462','HH1158'],
            'Nome':  ['Victor', 'Amanda', 'Bruna', 'Carlos', 'Ricardo', 'Maria'],
            'Idade': [20,35,40,54,30,27],
            'CEP': ['00092-029','11111-111','22222-888','00000-999','88888-111','77777-666']
           }

cadastro_a = pd.DataFrame(cadastro_a, columns = ['Id','Nome','Idade','CEP'])
cadastro_a
```

```python
# Cadastro da loja b
ID = ['CC9999','EF4488','DD9999','GT3462','HH1158']
name = ['Marcos', 'Patricia', 'Ericka', 'Ricardo', 'Maria',]
age = [19,30,22,30,27]
zip_code = ['00092-029','11111-111','11111-888','88888-111','77777-666']

informations = np.array([ID, name, age, zip_code])
informations_transpose = informations.T

cadastro_b = pd.DataFrame(informations_transpose, columns=['Id','Nome','Idade','CEP'])
cadastro_b
```

```python
# Registro de compras de todas unidades
compras = {'Id': ['AA2930','EF4488','CC2139','EF4488','CC9999','AA2930','HH1158','HH1158'],
            'Data':  ['2019-01-01','2019-01-30','2019-01-30','2019-02-01','2019-02-20','2019-03-15','2019-04-01','2019-04-10'],
            'Valor': [200,100,40,150,300,25,50,500]
           }
 
compras = pd.DataFrame(compras, columns = ['Id','Data','Valor'])
compras
```

# **Merge de dois DataFrame**

### **pd.merge(tabela_esquerda, tabela_direita, on = ['coluna_de_referência'], how = 'left/right/ inner/ outer')**

Representa uma estrutura que irá juntas duas tabelas através de uma coluna de referência. 

*on* = Qual coluna, coincidente em ambas as tabelas, será utilizada como referência. 

*how* = Forma como iremos concatená-las. 

--------------------

*left join* = Tabela principal será igual a esquerda. Nessa situação, os elementos da direita, que não contém as mesmas colunas dessa tabela, terão 'NaN' na posição.

*right join* =  Tabela da direita toda. 

*full join* = *outer join* = Junção de ambas as tabelas. 

*inner join* = Interceção.

```python
#left Join, usando como referência a coluna Id

merge_cadatro_a_compras = pd.merge(cadastro_a, compras, on=["Id"], how="left")
merge_cadatro_a_compras
```

## **Comentário sobre merge_cadatro_a_compras**

Note que, no DataFrame cadastro_a, não há as colunas: Data e Valor. As colunas de ambos os conjuntos entrarão na tabela, uma vez que fizemos o merge. No entanto, aqueles elementos da coluna ID (referência) que não apresentam colunas correspondentes a alguns dos DataFrames terão na sua linha 'NaN'. 

elementos "BB4563, DE2521 e GT3462" não estão na tabela "compras", logo não têm informações sobre Data e Valor. 


### **Observação:** Por que não escolhi outra coluna como referência?

Primeiro é necessário certificar se a coluna é existente nas duas tabelas que se deseja realizar o merge. 

Segundo, devemos analisar se não há elemento repetidos nas colunas. Casa haja, é valido realizar *pd.concat*

--------------------

## **Por que não usar a função concat?**

A função concat também serve para unir dois DataFrames, porém a função merge mescla as informações.

```python
pd.concat([cadastro_a, compras])
```

# **Introdução ao Groupby**

Se oberservarmos a variável *merge_cadatro_a_compras*, podemos ver que há algumas linhas com mesmos elementos da coluna de referência ("Id") duplicadas. 

Isso acontece, pois na coluna "Valor" temos duas linhas - para o mesmo elemento -  armazenando preços diferentes. 

Uma forma de mitigar tal empasse é através da soma dos preços, uma vez que eles pertencem a mesma pessoa. 

________________

## **nome_DataFrame.groupby(["coluna_de_referência"]).operação_que_deseja_fazer()**

```python
merge_cadatro_a_compras.groupby(['Valor']).sum()
```

```python
#Caso semelhante à célula acima, porém com as outras colunas. 
merge_cadatro_a_compras.groupby(['Id','Nome','Idade','CEP'])['Valor'].sum()
```

**Outros exemplos usando o groupby**

```python
df =  pd.DataFrame({'A': ['verdadeiro', 'verdadeiro', 'verdadeiro', 'falso', 
                          'falso', 'verdadeiro', 'falso', 'verdadeiro'], 
                    'B': ['um', 'dois', 'um', 'dois', 'três', 'dois', 'um', 'três'],
                    'C': np.random.rand(8),
                    'D': np.random.rand(8)})
df
```

```python
#Somar todos os elementos que posuem a mesma refência da coluna A

df.groupby(['A']).sum()
```

```python
#Processo mecânico da função groupby acima

truth_elements = 0
false_elements = 0

for i in range(len(df["A"])):
  if df["A"].tolist()[i] == "verdadeiro":
    truth_elements = truth_elements + df["C"].tolist()[i]
  elif df["A"].tolist()[i] == "falso":
    false_elements = false_elements + df["C"].tolist()[i]

pd.DataFrame([["falso", false_elements], ["verdadeiro", truth_elements]], columns=list("AC"))
```
